```python
import matplotlib.pyplot as plt
import pandas as pd 
import os 
from tqdm import tqdm
```

```python
plt.figure(figsize=(10, 10))
nrj = pd.read_csv('energy.xvg', sep='  ', header=None)
nrj.columns = ['time', 'potential', 'kinetic', 'total']
nrj.plot(x='time', y=['potential', 'kinetic', 'total'])
plt.title('Energy')
plt.xlabel('Time (ps)')
plt.ylabel('Energy (kJ/mol)')
plt.legend(['Potential', 'Kinetic', 'Total'])
plt.savefig('energy.png', dpi=500)
plt.show()
```

```python
temp = pd.read_csv('temperature.xvg', sep='  ', header=None)
temp.columns = ['time', 'temperature']
temp.plot(x='time', y='temperature')
plt.title('Temperature')
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.legend(['Temperature'])
plt.savefig('temperature.png', dpi=500)
plt.show()
```

```python
press = pd.read_csv('pression.xvg', sep='  ', header=None, index_col=False)
press.columns = ['time', 'pression']
press.plot(x='time', y='pression')
plt.title('Pression')
plt.xlabel('Time (ps)')
plt.ylabel('Pression (bar)')
plt.legend(['Pression'])
plt.savefig('pression.png', dpi=500)
plt.show()
press
```

```python
rmsf = pd.read_csv('rmsf.xvg', sep='  ', header=None, index_col=False)
rmsf.columns = ['residue', 'rmsf']
rmsf.plot(x='residue', y='rmsf')
plt.title('RMSF')
plt.xlabel('Residue')
plt.ylabel('RMSF (nm)')
plt.legend(['RMSF'])
plt.savefig('rmsf.png', dpi=500)
plt.show()
```
