```python
import pandas as pd 
```

```python
f = open("DOPE.log", "r")
line = f.readline()
data = {}
while line:
    DOPE = float(line)
    line = f.readline()
    GA = float(line)
    line = f.readline()
    name = line.strip()
    data[name] = [DOPE, GA]
    line = f.readline()
```

```python
df = pd.DataFrame.from_dict(data, orient='index', columns=['DOPE', 'GA'])
df.sort_values(by=['DOPE', 'GA'], ascending=[True, False], inplace=True)
df.head(20)
```
